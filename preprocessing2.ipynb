```python
import ast
```

```python
# Method One
for i in range(data.shape[0]):
    data['genres'][i] = ast.literal_eval(data['genres'][i])
```

```python
# Method Two
data['cast'] = data['cast'].map(lambda i : ast.literal_eval(i))
data['crew'] = data['crew'].map(lambda i : ast.literal_eval(i))
```

```python
# Get a list of all genres
```

```python
data['genres'][2]
```

```python
def getgenres(x):
    gen=[]
    for i in x:
        if (i.get('name')=='Science Fiction'):
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
            
    if (gen==[]):
        return np.NaN
    else:
        gen = ' '.join(gen)
        return gen
```

```python
data['genres'] = data['genres'].map(lambda i : getgenres(i))
```

```python
data['genres'][0]
```

```python
# Get a column for actor_1_name
```

```python
data['cast'][0]
```

```python
data['cast'][0][0].get('name')
```

```python
def actor1(x):
    actor_1_name = []
    for i in x:
        actor_1_name.append(i.get('name'))
    if actor_1_name==[]:
        return np.NaN
    else:
        return actor_1_name[0]
```

```python
data['actor_1_name'] = data['cast'].map(lambda i : actor1(i))
```

```python
data['actor_1_name'][0]
```

```python
# Get a column for actor_2_name and actor_3_name
```

```python
def actor2(x):
    actor_2_name = []
    for i in x:
        actor_2_name.append(i.get('name'))
    if actor_2_name==[] or len(actor_2_name)<=1:
        return np.NaN
    else:
        return actor_2_name[1]
```

```python
data['actor_2_name'] = data['cast'].map(lambda i : actor2(i))
```

```python
def actor3(x):
    actor_3_name = []
    for i in x:
        actor_3_name.append(i.get('name'))
    if actor_3_name==[] or len(actor_3_name)<=2:
        return np.NaN
    else:
        return actor_3_name[2]
```

```python
data['actor_3_name'] = data['cast'].map(lambda i : actor3(i))
```

```python
data.head()
```

```python
# Get a column for director_name
```

```python
data['crew'][0]
```

```python
def director(x):
    dl = []
    st =" "
    for i in x:
        if (i.get('job')=='Director'):
            dl.append(i.get('name'))
    if dl==[]:
        return np.NaN
    else:
        return st.join(dl)
```

```python
data['director_name'] = data['crew'].map(lambda i : director(i))
```
